# Ex 4: Daily Maths

You are working in a retail shop. Write a program that computes the total cost for a customer that bought 3 differnt types of fruits : apple, orange, mango. Ask the user to enter the quantity and price for each type.

```python
apple_qty = int(input("Enter quantity of apples: "))
apple_unitprice = float(input("Enter unit price of apple: "))

orange_qty = int(input("Enter quantity of oranges: "))
orange_unitprice = float(input("Enter unit price of orange: "))

mango_qty = int(input("Enter quantity of mangoes: "))
mango_unitprice = float(input("Enter unit price of mango: "))

apple_total = apple_qty * apple_unitprice
orange_total = orange_qty * orange_unitprice
mango_total = mango_qty * mango_unitprice

grand_total = apple_total + orange_total + mango_total

print("Total cost of apples: {}".format(apple_total))
print("Total cost of oranges:{}".format(orange_total))
print("Total cost of mangoes:{}".format(mango_total))
print("Grand Total Cost:{}".format(grand_total))
```

# Ex 9: Computing

Write a program that asks the user for a number of seconds and prints out how many minutes and seconds that is. For instance, 200 seconds is 3 minutes and 20 seconds.

```python
input_sec = int (input("Enter number of seconds: "))
minutes = input_sec // 60      #Use // Integer division gives minutes
seconds = input_sec % 60       #Remainder gives seconds
print("{} seconds is {} minutes and {} seconds".format(input_sec,minutes,seconds))
```
